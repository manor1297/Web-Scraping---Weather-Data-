```python
#importing requiredd libraries

import numpy as np
import pandas as pd
import requests
from bs4 import BeautifulSoup
import lxml
import re

#creating the DateTimeIndex for indexes of dataframe

columns_array = ['Average temperature (°F)', 'Average humidity (%)',
 'Average dewpoint (°F)', 'Average barometer (in)',
 'Average windspeed (mph)', 'Average gustspeed (mph)',
 'Average direction (°deg)', 'Rainfall for month (in)',
 'Rainfall for year (in)', 'Maximum rain per minute',
 'Maximum temperature (°F)', 'Minimum temperature (°F)',
 'Maximum humidity (%)', 'Minimum humidity (%)', 'Maximum pressure',
 'Minimum pressure', 'Maximum windspeed (mph)',
 'Maximum gust speed (mph)', 'Maximum heat index (°F)']

date_iter = pd.date_range(start='01/01/2009', end='28/11/2018', freq='M')
day_iter = pd.date_range(start='01/01/2009', end='28/10/2018', freq='D')
DATE_TIME_INDEX = np.empty([0],dtype='<M8[ns]')

#creating arrays
AVG_TEMP = np.empty([0], dtype=float)
AVG_HUM = np.empty([0], dtype=int)
AVG_DEW = np.empty([0], dtype=float)
AVG_BAR = np.empty([0], dtype=float)
AVG_WIND = np.empty([0], dtype=float)
AVG_GUST = np.empty([0], dtype=float)
AVG_DIR = np.empty([0], dtype=int)
RAIN_MONTH = np.empty([0], dtype=float)
RAIN_YEAR = np.empty([0], dtype=float)
MAX_RPM = np.empty([0], dtype=float)
MAX_TEMP = np.empty([0], dtype=float)
MIN_TEMP = np.empty([0], dtype=float)
MAX_HUM = np.empty([0], dtype=int)
MIN_HUM = np.empty([0], dtype=float)
MAX_PR = np.empty([0], dtype=float)
MIN_PR = np.empty([0], dtype=float)
MAX_WIND = np.empty([0], dtype=float)
MAX_GUST = np.empty([0], dtype=float)
MAX_HI = np.empty([0], dtype=float)

data_array =  [AVG_TEMP, AVG_HUM, AVG_DEW, AVG_BAR, AVG_WIND, AVG_GUST, AVG_DIR, RAIN_MONTH, RAIN_YEAR, MAX_RPM, MAX_TEMP,
               MIN_TEMP, MAX_HUM, MIN_HUM, MAX_PR, MIN_PR,MAX_WIND, MAX_GUST, MAX_HI]

#creating iterators for url date and month manipulation
months_date_iter = ["{:02d}".format(i.month) for i in date_iter]
year_date_iter = ["{}".format(i.year) for i in date_iter]
```

```python
for i,j in zip(year_date_iter,months_date_iter):
    #print("Year - {} and month - {}\n".format(i,j))
    #Making request variable
    res = requests.get('http://www.estesparkweather.net/archive_reports.php?date={}{}'.format(i,j))
    
    #making bs4 variable
    soup = BeautifulSoup(res.text, 'lxml')
    soup.text
    
    #Getting all table tags
    td_tags = soup.select('td')
    td_tags_text = np.array([i.text for i in td_tags])
    
    #Creating arrays for each column
    Avg_temp = td_tags_text[np.array([i+1 for i in range(len(td_tags_text)) if td_tags_text[i].startswith('Average temperature')])]
    Avg_temp = np.char.strip(Avg_temp, ' °F').astype(float)
    Avg_hum = td_tags_text[np.array([i+1 for i in range(len(td_tags_text)) if td_tags_text[i].startswith('Average humidity')])]
    Avg_hum = np.char.strip(Avg_hum, ' %').astype(int)
    Avg_dew = td_tags_text[np.array([i+1 for i in range(len(td_tags_text)) if td_tags_text[i].startswith('Average dewpoint')])]
    Avg_dew = np.char.strip(Avg_dew, ' °F').astype(float)
    Avg_bar = td_tags_text[np.array([i+1 for i in range(len(td_tags_text)) if td_tags_text[i].startswith('Average barometer')])]
    Avg_bar = np.char.strip(Avg_bar, ' in.').astype(float)
    Avg_wind = td_tags_text[np.array([i+1 for i in range(len(td_tags_text)) if td_tags_text[i].startswith('Average windspeed')])]
    Avg_wind = np.char.strip(Avg_wind, ' mph').astype(float)
    Avg_gust = td_tags_text[np.array([i+1 for i in range(len(td_tags_text)) if td_tags_text[i].startswith('Average gustspeed')])]
    Avg_gust = np.char.strip(Avg_gust, ' mph').astype(float)
    Avg_dir = td_tags_text[np.array([i+1 for i in range(len(td_tags_text)) if td_tags_text[i].startswith('Average direction')])]
    Avg_dir = np.char.strip(Avg_dir, ' ° (WSNE)').astype(int)
    Rain_month = td_tags_text[np.array([i+1 for i in range(len(td_tags_text)) if td_tags_text[i].startswith('Rainfall for month')])]
    Rain_month = np.char.strip(Rain_month, ' in.').astype(float)
    Rain_year = td_tags_text[np.array([i+1 for i in range(len(td_tags_text)) if td_tags_text[i].startswith('Rainfall for year')])]
    Rain_year = np.char.strip(Rain_year, ' in.').astype(float)
    Max_rpm = td_tags_text[np.array([i+1 for i in range(len(td_tags_text)) if td_tags_text[i].startswith('Maximum rain per min')])]
    Max_rpm = np.array([i.split(" ")[1] for i in Max_rpm]).astype(float)
    Max_temp = td_tags_text[np.array([i+1 for i in range(len(td_tags_text)) if td_tags_text[i].startswith('Maximum temperature')])]
    Max_temp = np.array([i.split(" ")[1] for i in Max_temp])
    Max_temp = np.char.strip(Max_temp, ' °F').astype(float)
    Min_temp = td_tags_text[np.array([i+1 for i in range(len(td_tags_text)) if td_tags_text[i].startswith('Minimum temperature')])]
    Min_temp = np.array([i.split(" ")[1] for i in Min_temp])
    Min_temp = np.char.strip(Min_temp, ' °F').astype(float)
    Max_hum = td_tags_text[np.array([i+1 for i in range(len(td_tags_text)) if td_tags_text[i].startswith('Maximum humidity')])]
    Max_hum = np.array([re.search(r'\d+', i).group() for i in Max_hum])
    Max_hum = Max_hum.astype(int)
    Min_hum = td_tags_text[np.array([i+1 for i in range(len(td_tags_text)) if td_tags_text[i].startswith('Minimum humidity')])]
    Min_hum = np.array([re.search(r'\d+', i).group() for i in Min_hum])
    Min_hum = Min_hum.astype(int)
    Max_pr = td_tags_text[np.array([i+1 for i in range(len(td_tags_text)) if td_tags_text[i].startswith('Maximum pressure')])]
    Max_pr = np.array([i.split(" ")[1] for i in Max_pr]).astype(float)
    Min_pr = td_tags_text[np.array([i+1 for i in range(len(td_tags_text)) if td_tags_text[i].startswith('Minimum pressure')])]
    Min_pr = np.array([i.split(" ")[1] for i in Min_pr]).astype(float)
    Max_wind = td_tags_text[np.array([i+1 for i in range(len(td_tags_text)) if td_tags_text[i].startswith('Maximum windspeed')])]
    Max_wind = np.array([i.split(" ")[1] for i in Max_wind]).astype(float)
    Max_gust = td_tags_text[np.array([i+1 for i in range(len(td_tags_text)) if td_tags_text[i].startswith('Maximum gust speed')])]
    Max_gust = np.array([re.search(r'\d+\.?\d+', i).group() for i in Max_gust]).astype(float)
    Max_HI = td_tags_text[np.array([i+1 for i in range(len(td_tags_text)) if td_tags_text[i].startswith('Maximum heat index')])]
    Max_HI = np.array([re.search(r'\d+\.?\d+', i).group() for i in Max_HI])
    Max_HI = Max_HI.astype(float)
    
    #Getting the missing days
    td_tags_text_month = ["{}{}".format(re.search(r'\w+\s\d+',x).group().replace(" ", ""),i) 
                          for x in td_tags_text if x.startswith(('Jan','Feb','Mar','Apr',
                                                                'May','Jun','Jul','Aug',
                                                                'Sep','Oct','Nov','Dec'))]
    date_time_index = np.array(pd.to_datetime(td_tags_text_month, format="%b%d%Y"))
    DATE_TIME_INDEX = np.hstack((DATE_TIME_INDEX,date_time_index))
    
    tdata_array = [Avg_temp, Avg_hum, Avg_dew, Avg_bar, Avg_wind, Avg_gust, Avg_dir, Rain_month, Rain_year, Max_rpm, Max_temp,
               Min_temp, Max_hum, Min_hum, Max_pr, Min_pr,Max_wind, Max_gust, Max_HI]
    
    #appending arrays
    for i in range(len(data_array)):
        data_array[i] = np.hstack((data_array[i], tdata_array[i][:-1]))
```

```python
#datetimeindex of days excluding missing days
DATE_TIME_INDEX = np.array([ i[:10] for i in DATE_TIME_INDEX.astype(str) ])
```

```python
#creating the data frame
df = pd.DataFrame(columns=columns_array, index=day_iter)
k = 0
for i in DATE_TIME_INDEX[:-3]:
    for j in range(19):
        df.loc[i][j] = pd.to_numeric(data_array[j][k])
    k += 1
```
